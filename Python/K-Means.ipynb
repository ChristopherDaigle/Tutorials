# Unsupervised learning method
We start with a generated data set that we know the "right" attributes of (the response) and test it.

We then skip forward to a real data set where the "right" attributes are unknown. Thus, the algorithm can help us cluster for an unsupervised problem.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
df = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
df[0].shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df[0][:,0], df[0][:,1], c = df[1], cmap = 'rainbow')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (10,6))

ax1.set_title('K Means')
ax1.scatter(df[0][:,0], df[0][:,1], c = kmeans.labels_, cmap = 'rainbow')

ax2.set_title('Original')
ax2.scatter(df[0][:,0], df[0][:,1], c = df[1], cmap = 'rainbow')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df[0])

fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (10,6))

ax1.set_title('K Means')
ax1.scatter(df[0][:,0], df[0][:,1], c = kmeans.labels_, cmap = 'rainbow')

ax2.set_title('Original')
ax2.scatter(df[0][:,0], df[0][:,1], c = df[1], cmap = 'rainbow')
plt.show()
```
