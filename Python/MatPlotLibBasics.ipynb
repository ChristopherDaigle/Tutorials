```python
# Multiplot on the same canvas
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
#plt.plot(x, y, 'bo')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# OBJECT ORIENTED (OO) METHOD
fig = plt.figure()

#[left, bottom, width, heighth]; width and heighth range from [0,1], percent of blank canvas
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r+')
axes1.set_title('LARGER PLOT')
axes1.set_xlabel('X1')
axes1.set_ylabel('Y1')

axes2.plot(y, x, 'bo')
axes2.set_title('SMALLER PLOT')
axes2.set_xlabel('X2')
axes2.set_ylabel('Y2')
```

```python
# Create an empty figure
fig = plt.figure()
# add a set of axes to the blank canvis
axes1 = fig.add_axes([0.0,0.0,0.7,0.7])
# Create data
x = np.linspace(0,5,11)
y = x ** 2
# plot the data on the axes
axes1.plot(x, y)
```

```python
## The second lecture
fig,axes = plt.subplots(nrows = 1, ncols = 2)
# plt.tight_layout()
#axes.plot(x, y)
##
## Iterate through the axes I defined earlier is possible now
# for current_ax in axes:
#     current_ax.plot(x, y)
## These axes come from where I previously defined them
## I can select them by indexing as well
axes[0].plot(x, y, 'r-')
axes[0].plot(x, y, 'b*')
axes[0].set_title('First Plot')

axes[1].plot(y, x, 'r-')
axes[1].plot(y, x, 'bo')
axes[1].set_title('Second Plot')

plt.tight_layout()
```

# Figure Size and DPI
